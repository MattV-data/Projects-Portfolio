## Conclusions

#### During the timespan, it was seen that Thursdays were the days with the most Uber rides, followed Wednesday and then Friday.

#### During weekdays, there was a spike in ridership at around 7-8am (morning rush hour), with a maximum peak around 5-6pm (after work rush hour). There was a additional peak, for Friday and Saturday only at around 11pm.

#### During periods of extreme temperatures, requests for Uber rides are much higher than normal. Additionally, during periods of excessive precipitation, requests for Rides are again higher than normal.

#### An interesting observation, during the monthly ridership, Wednesday seemed to be the day with the most rides in the 4 month window, (coming in number 1 in 2 months and a close second in a third. However, in the month of May, Thursday and Friday surpase Wednesday for rides. It has been hypothesized that with tempuratures rising in May, with the end of winter, onset of Spring and soon summer, many people may have visited bars/pubs after work. Those who may have normally taken public transport instead chose Uber rides later in the evening.
